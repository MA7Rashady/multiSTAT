## 3.2 Geometry of the Sample

A single multivariate observation is the collection of measurements on p different variables taken on the same item or trial.

If n observations have been obtained, the entire data set can be placed in an n × p matrix:

>Remark. We often say that the data are a sample of size n from a p-variate population.
The sample then consists of n measurements, each of which has p components.

The data can be plotted in two different ways:

* For the p-dimensional scatterplot, the rows of X represent n points in p-dimensional space.

The scatterplot matrix of n points in p-dimensional space provides information on
the locations and variability of the points.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[4, 1], [-1, 3], [3, 5]])

# each point represent an observation

plt.scatter(X[:,0], X[:,1],s=150 ,c='r')
plt.scatter(X.mean(0)[0], X.mean(0)[1], marker="*",s=450 ,c='g')
plt.grid()
```

* Consider the data as p vectors in n-dimensional space. Here we take the elements
of the columns of the data matrix to be the coordinates of the vectors:

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

X = np.array([[4, 1], [-1, 3], [3, 5]])  # X Matrix

fig = plt.figure(figsize=(10, 8))
ax = fig.gca(projection='3d')

# each vector represent a variable
ax.quiver(0,0,0, 4,-1,3)
ax.quiver(0,0,0, 1, 5,5)

ax.set_xlim([0,5])
ax.set_ylim([0,5])
ax.set_zlim([0,5])
ax.set_xlabel('Observ 1')
ax.set_ylabel('Observ 2')
ax.set_zlabel('Observ 3')
plt.show()
```
